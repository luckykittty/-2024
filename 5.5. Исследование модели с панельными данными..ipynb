```python
import pandas as pd
from linearmodels.panel import PanelOLS

# Загрузка данных
data = pd.read_excel('Панельные_данные.xlsx')
```

```python
data
```

```python

# Создание панельных данных
Panel_data = data.set_index(['N', 'G'])

# Построение модели
model = PanelOLS.from_formula('Y ~ 1 + X1 + X2 ', data=Panel_data)
result = model.fit()

# Вывод результатов
print(result)
```

```python

# Создание панельных данных
Panel_data = data.set_index(['N', 'G'])

# Построение модели
model = PanelOLS.from_formula('Y ~ X1 + X2 + EntityEffects', data=Panel_data)
result = model.fit(cov_type='clustered', cluster_entity=True)

# Вывод результатов
print(result)
```
