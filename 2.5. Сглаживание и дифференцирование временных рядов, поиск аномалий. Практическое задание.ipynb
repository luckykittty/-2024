```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
X = np.arange(10,21)
Y = np.array([12, 15, 18, 16, 24, 22, 27, 28, 25, 32, 28])
print('mean X =', X.mean(), ', mean Y', Y.mean())
print(f'std(X) = {X.std(ddof=1)}, std(Y) = {Y.std(ddof=1)}, Var(X) = {X.var(ddof=1)}, Var(Y) = {Y.var(ddof=1)}')
print('Cov(X, Y) =', np.cov(X, Y)[0,1])
print('Cov(X, Y) =', np.corrcoef(X, Y)[0,1])
plt.scatter(X, Y)
plt.title('Коррелограмма X и Y')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```
